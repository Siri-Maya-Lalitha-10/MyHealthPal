```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pdf=pd.read_csv('parkinsons.csv')
```

```python
pdf.head()
```

```python
print("Total missing values in Parkinson's data:", pdf.isnull().sum().sum())
```

```python
# Basic statistics and distribution
print(pdf.describe())
```

```python
# Convert the 'status' column as the target variable and drop the 'name' column
X = pdf.drop(columns=['status', 'name'])
y = pdf['status']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=2)
```

```python
# Train SVM model
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Evaluate the model
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print("Training Accuracy:", train_score)
print("Test Accuracy:", test_score)
```

```python
# Display classification report
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
